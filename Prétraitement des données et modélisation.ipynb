# Imputation

```python
def impute_age(cols):
    age=cols[0]
    pclass=cols[1]
    
    if pd.isnull(age):
        if pclass==1:
            return 37.0000
        elif pclass==2:
            return 29.0000
        else :
            return 24.0000
    else:
        return age
df['age']=df[['age','pclass']].apply(impute_age,axis=1)
df
```

# Standardisation

```python
def standardisation(df):
    quantitative_vars = df.select_dtypes(include=['float64','int64'])
    qualitative_vars = df.select_dtypes(include=['object'])
    scaler = StandardScaler()
    scaled_quantitative_vars = pd.DataFrame(scaler.fit_transform(quantitative_vars), columns=quantitative_vars.columns)
    df = pd.concat([scaled_quantitative_vars, qualitative_vars], axis=1)
    return df
df = standardisation(df)
df
```

# Train_set,Test_set,Nettoyage,Encodage

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

# Train_set, Test_set

```python
trainset, testset = train_test_split(df,test_size=0.2,random_state=0)
```

# Encodage et Nettoyage

```python
def encodage(df):
    code = {'Survecu':0,
            'Decede':1,
            'male':0,
            'female':1,
            'classe1':0,
            'classe2':1,
            'classe3':2,
            'S':0,
            'C':1,
            'Q':2
           }
    for col in df.select_dtypes('object').columns:
        df[col] = df[col].map(code)
        
    return df
```

```python
def imputation(df):
    df = df.dropna(axis=0)
    return df
```

```python
def preprocessing(df):
    df = encodage(df)
    df = imputation(df)
    
    x = df.drop('survived',axis=1)
    y = df['survived']
    print(y.value_counts())
    return(x,y)
```

```python
x_train, y_train = preprocessing(trainset)
```

```python
x_test, y_test= preprocessing(testset)
```
